```python
# PRINT 7 - Nivel 1 - Ejercicio 1: Calculadora de IMC 

def intro_peso():
    
    print("SPRINT 7 - Nivel 1 Calculadora de IMC")
    print("Bienvenido a la calculadora de IMC")
    print("El índice de masa corporal (IMC) es una forma de medir el tamaño corporal. "
          "Es una herramienta para estimar la grasa corporal y detectar la obesidad y los riesgos para la salud")
    print("Para calcular tu IMC, introduce tu peso en kilogramos y tu altura en centímetros.")

def pedir_datos_usuario():
  
    peso = float(input("Introduce tu peso en Kg: "))
    altura = int(input("Introduce tu altura en cm: "))
    return peso, altura

def calcular_imc(peso, altura_cm):

    altura_metros = altura_cm / 100
    imc = peso / (altura_metros ** 2)
    return imc

def clasificar_imc(imc):
    
    if imc < 18.5: 
        return "Tienes Bajo peso"
    elif 18.5 <= imc < 24.9:
        return "Tienes un peso saludable"
    elif 25 <= imc < 29.9:
        return "Tienes Sobrepeso"
    elif 30 <= imc < 34.9:
        return "Tienes Obesidad grado 1 (moderada)"
    elif 35 <= imc < 39.9:
        return "Tienes Obesidad grado 2 (severa)"
    else:
        return "Tienes Obesidad grado 3 (muy severa o mórbida)"

def main():
    intro_peso()
    peso, altura = pedir_datos_usuario()
    imc = calcular_imc(peso, altura)
    print(f"Basado en tu peso ({peso} kg) y tu altura ({altura} cm), tu IMC es: {imc:.2f}")
    print(clasificar_imc(imc))
    print("Gracias por usar la calculadora de IMC. ¡Cuídate!")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```

```python
# SPRINT 7 - Nivel 1 - Ejercicio 2: Conversor de temperatura
# Conversor de temperatura Celsius a Fahrenheit y Kelvin y viceversa
# Se utilizara def para definir las intro y las unidades de temperatura
# Se utilizará while True , Input y break para que el usuario pueda elegir la temperatura a convertir y el tipo de conversión y poder repetir el proceso
# Se utilizará if y elif para definir el tipo de conversión y las fórmulas de conversión
```

```python
def intro_temperatura():
    print("SPRINT 7 - Nivel 1 Conversor de temperatura")
    print("Bienvenido al conversor de temperatura")
    print("Selecciona la unidad de temperatura que deseas comparar:")
    print("1. Celsius (°C)")
    print("2. Fahrenheit (°F)")
    print("3. Kelvin (K)")

def celsius_a_otras(temp):
    fahrenheit = (temp * 9/5) + 32
    kelvin = temp + 273.15
    return f"{temp:.2f} °C (Celsius) son {fahrenheit:.2f} °F (Fahrenheit) y {kelvin:.2f} K (Kelvin)"

def fahrenheit_a_otras(temp):
    celsius = (temp - 32) * 5/9
    kelvin = celsius + 273.15
    return f"{temp:.2f} °F (Fahrenheit) son {celsius:.2f} °C (Celsius) y {kelvin:.2f} K (Kelvin)"

def kelvin_a_otras(temp):
    celsius = temp - 273.15
    fahrenheit = (celsius * 9/5) + 32
    return f"{temp:.2f} K (Kelvin) son {celsius:.2f} °C (Celsius) y {fahrenheit:.2f} °F (Fahrenheit)"

def nueva_consulta():
    respuesta = input("¿Deseas realizar otra consulta? (si/no): ").strip().lower()
    return respuesta == "si"

def conversor_temperatura():
    while True:
        intro_temperatura()
        opcion = int(input("Introduce el número de la opción que deseas comparar: "))
        temperatura = float(input("Introduce la temperatura: "))

        if opcion == 1:
            resultado = celsius_a_otras(temperatura)
            print(resultado)
        elif opcion == 2:
            resultado = fahrenheit_a_otras(temperatura)
            print(resultado)
        elif opcion == 3:
            resultado = kelvin_a_otras(temperatura)
            print(resultado)
        else:
            print("Opción no válida. Por favor, selecciona una opción entre 1 y 3.")

        if not nueva_consulta():
            print("Gracias por usar este increíble conversor de temperatura. ¡Hasta luego!")
            break

# Ejecutar el conversor
conversor_temperatura()
```

```python
# SPRINT 7 - Nivel 1 - Ejercicio 3: Contador de palabras en un texto 
# Se importara  la librería re para trabajar con expresiones regulares (Caracteres especiales)
# Se utilizara la funcion input para que el usuario pueda introducir el texto a analizar
# Se utilizara la funcion for para ejecutar el bucle y contar las palabras
# Se utilizara la funcion split para eliminar los espacios en blanco y dividir el texto en palabras
# Se utilizara la funcion lower para convertir el texto a minusculas y evitar duplicados
# Se utilizara la funcion get para obtener el valor de la palabra y contar las veces que aparece en el texto
# Se utilizara la funcion items para mostrar el resultado de las palabras y su cuenta
# se utilizara la funcion sorted para ordenar el resultado de las palabras y su cuenta
# Se utilizara while True y break para que el usuario pueda introudir un nuevo texto despues de que se le muestre el resultado y finalizar el programa en caso de que el usuario no quiera introducir un nuevo texto
```

```python
# SPRINT 7 - Nivel 1 - Ejercicio 3:
import re  

def intro_contador():
    print("SPRINT 7 - Nivel 1 contador de palabras en un texto")
    print("Bienvenido al contador de palabras")
    print("Este programa cuenta la cantidad de veces que aparece cada palabra en un texto.")
    print("Por favor, introduce a continuación el texto a analizar:")

intro_contador()  # Llamar a la función de introducción

def contador_palabras():
    while True:
        texto = input("Introduce el texto a analizar: ")  # Solicitar al usuario el texto a analizar
        texto = texto.lower()  # Convertir el texto a minúsculas para evitar duplicados
        palabras = re.findall(r'\b[\w@.]+\b', texto)  # Usar una expresión regular para extraer palabras que incluyan letras, números y caracteres como '@' o '.'
        
        cuenta_palabra = {}  # Crear un diccionario para almacenar la cuenta de palabras
        
        for palabra in palabras:
            # Incrementar el conteo de cada palabra
            cuenta_palabra[palabra] = cuenta_palabra.get(palabra, 0) + 1
        
        # Ordenar el conteo de palabras de mayor a menor
        palabras_ordenadas = sorted(cuenta_palabra.items(), key=lambda x: x[1], reverse=True)
        
        # Mostrar el conteo de palabras
        print("Conteo de palabras (ordenado de mayor a menor):")
        for palabra, count in palabras_ordenadas:
            print(f"'{palabra}': {count} veces")
        
        # Preguntar si desea realizar otra consulta
        nueva_consulta = input("¿Deseas analizar otro texto? (si/no): ").strip().lower()
        if nueva_consulta != "si":
            print("Gracias por usar el contador de palabras. ¡Hasta luego!")
            break

contador_palabras()  # Llamar a la función contador_palabras
```

```python
def reverse_dictionary(input_dict):
    reversed_dict = {}
    seen_values = {}
    for key, value in input_dict.items():
        if value in seen_values:
            print("Error: múltiples claves para un mismo valor")
            return None
        reversed_dict[value] = key
        seen_values[value] = True
    return reversed_dict

# Ejecucion del diccionario 
dict1 = {'a': 1, 'b': 2, 'c': 3}
reversed_dict1 = reverse_dictionary(dict1)
if reversed_dict1:
    print(reversed_dict1)  # Salida: {1: 'a', 2: 'b', 3: 'c'}

dict2 = {'x': 'manzana', 'y': 'plátano', 'z': 'plátano'}
reversed_dict2 = reverse_dictionary(dict2)
if reversed_dict2:
    print(reversed_dict2)  # Esto no se imprimirá porque la función retorna None
```

```python
def reverse_dictionary(input_dict):
    reversed_dict = {}
    for key, value in input_dict.items():
        if value in reversed_dict:
            # Si ya existe, convierte a lista si no lo es y añade el nuevo key
            if isinstance(reversed_dict[value], list):
                reversed_dict[value].append(key)
            else:
                reversed_dict[value] = [reversed_dict[value], key]
        else:
            reversed_dict[value] = key
    return reversed_dict

# Ejemplo de uso:
dict1 = {'a': 1, 'b': 2, 'c': 3}
print(reverse_dictionary(dict1))  # {1: 'a', 2: 'b', 3: 'c'}

dict2 = {'x': 'manzana', 'y': 'plátano', 'z': 'plátano'}
print(reverse_dictionary(dict2))  # {'manzana': 'x', 'plátano': ['y', 'z']}
```

```python
def conversion(data):
    def flatten(lst):
        for item in lst:
            if isinstance(item, (list, tuple)):
                yield from flatten(item)
            else:
                yield item

    floatables = []
    not_floatables = []
    for item in flatten(data):
        try:
            floatables.append(float(item))
        except (ValueError, TypeError):
            not_floatables.append(item)
    return floatables, not_floatables

# Ejemplo de uso:
entrada = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]
print(conversion(entrada))
```

```python
# SPRINT 7 - Nivel 3 - Ejercicio 1: Contador de palabras en un texto 
# Se utilizara import re para trabajar con expresiones regulares
# Se utilizara import os para trabajar con el sistema operativo
# Se creará una función llamada intro_contador para mostrar la introducción del programa
# Se creará una función llamada leer_archivo para leer el contenido de un archivo de texto y retornarlo como una cadena
#    usamos os.path.isfile para verificar si el archivo existe
#    Se usa  encoding='utf-8' para manejar caracteres especiales
# Se creará una función llamada contador_palabras para contar las palabras en el texto leído
#   Se utilizará un diccionario para almacenar las palabras y su conteo
#   Se utilizará main
```

```python
import re
import os

def intro_contador():
    print("SPRINT 7 - Nivel 3 - Ejercicio 1: Contador de palabras en archivo TXT")
    print("Bienvenido al contador de palabras por archivo")
    print("Este programa lee un archivo TXT, cuenta la frecuencia de cada palabra y las organiza alfabéticamente por letra inicial.")

def leer_archivo_txt(ruta_archivo):
    if not os.path.isfile(ruta_archivo):
        print(f"El archivo no existe o la ruta es incorrecta:\n{ruta_archivo}")
        return ""
    with open(ruta_archivo, "r", encoding="utf-8") as f:
        return f.read()

def contar_palabras(texto):
    texto = texto.lower()
    palabras = re.findall(r'\b[\w@.]+\b', texto)
    diccionario = {}
    for palabra in palabras:
        inicial = palabra[0].lower()
        if inicial not in diccionario:
            diccionario[inicial] = {}
        diccionario[inicial][palabra] = diccionario[inicial].get(palabra, 0) + 1
    diccionario_ordenado = {}
    for letra in sorted(diccionario.keys()):
        diccionario_ordenado[letra] = dict(sorted(diccionario[letra].items()))
    return diccionario_ordenado

def mostrar_diccionario(diccionario):
    for letra in diccionario:
        print(f"'{letra}': {{", end='') 
        palabras = list(diccionario[letra].items())
        for i, (palabra, count) in enumerate(palabras):
            coma = ',' if i < len(palabras) - 1 else ''
            print(f" '{palabra}': {count}{coma}", end='')
        print(" },")  

def main():
    intro_contador()
    ruta = input("Introduce la ruta completa del archivo TXT a analizar: ").strip()
    ruta = ruta.strip('"').strip("'")
    texto = leer_archivo_txt(ruta)
    if texto:
        diccionario_palabras = contar_palabras(texto)
        print("{")
        mostrar_diccionario(diccionario_palabras)
        print("}")

if __name__ == "__main__":
    main()
```
